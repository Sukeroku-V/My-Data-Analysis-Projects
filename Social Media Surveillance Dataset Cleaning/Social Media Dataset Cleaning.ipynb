```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv(r'C:\Users\User\Downloads\archive\Social Media Sureveillance Dataset.csv')
df
```

```python
#Replace unnecessary ';' with ','
df.loc[0, 'Authors'] = df.loc[0, 'Authors'].replace(';', ',')
df.loc[1, 'Author full names'] = df.loc[1, 'Author full names'].replace(';', ',')
df
```

```python
#missing data
for col in df.columns:
    missing_data = np.mean(df[col].isnull())
    print('{}, {} %'.format(col, missing_data))
```

```python
#remove duplicates
df = df.drop_duplicates()
df
```

```python
#fix data types for a direct use
df['Volume'] = pd.to_numeric(df['Volume'], errors = 'coerce')
df['Issue'] = pd.to_numeric(df['Issue'], errors = 'coerce')
df['Page count'] = pd.to_numeric(df['Page count'], errors = 'coerce')
df['Cited by'] = pd.to_numeric(df['Cited by'], errors = 'coerce')
```

```python
#Outliers
df['Volume'].describe()
```

```python
df.boxplot(column= 'Volume')
plt.show()
```

```python
#Correction of inconsistent data
df['Index Keyowrds'] = df['Index Keywords'].str.lower()
```

```python
#Replacing Blank rows
df.replace('', None, inplace=True)
df
```

```python
#Dropping unnecessary columns
df = df.drop(columns = ['ISSN', 'ISBN', 'CODEN', 'EID'])
df
```
