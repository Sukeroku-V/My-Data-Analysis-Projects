```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'c:\Users\User\Downloads\video_games_sales.csv\video_games_sales.csv')
df.head(5)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df = df.dropna()
```

```python
df.describe()
```

```python
# a bar chart for most commone genres
genre_counts = df['genre'].value_counts()
genre_counts.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(8, 5))
plt.title('Number of Video Games per Genre', fontsize=16)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
#10 most sucessful games
top_10_games = df.sort_values(by='global_sales', ascending=False).head(10)
top_10_games.plot(kind='barh', x='name', y='global_sales', color='skyblue', edgecolor='black', figsize=(10, 6))
plt.title('Top 10 Video Games by Global Sales', fontsize=16)
plt.xlabel('Global Sales (millions)', fontsize=14)
plt.ylabel('Game Name', fontsize=14)
plt.show()
```

```python
#partnered publishers
Top_5_publishers = df['publisher'].value_counts().head(5)
Top_5_publishers.plot(kind='barh', color='skyblue', edgecolor='black', figsize=(8, 5))
plt.title('Top 5 publishers', fontsize=16)
plt.xlabel('Publisher', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
#Platforms
Popular_platforms = df['platform'].value_counts().head(8)
Popular_platforms.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(8, 5))
plt.title('Popular Platforms', fontsize=16)
plt.xlabel('Platform', fontsize=14)
plt.ylabel('Players', fontsize=14)
plt.show()
```
